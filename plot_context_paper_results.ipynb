```python
dd = {}
for k1, v1 in data.items():
    for k2, v2 in v1.items():
        dd[(k1, k2)] = v2
#dd
```

```python
dd = {
    "index": dd.keys(),
    "columns": ["SIC", "SOD", "Floe", "Combined"],
    "data": dd.values(),
    "index_names": ["Feature Group", "Atrous Rate"],
    "column_names": ["Task"]
}
```

```python
dd["index"]
```

```python
df = pd.DataFrame.from_dict(dd, orient="tight")
df
```

```python
xlabels = [
    "AMSR2 (18.7, 36.5)", 
    "AMSR2 (18.7, 36.5, 23.8)", 
    "AMSR2 (All)", 
    "S1", 
    "S1 + AMSR2 (18.7, 36.5)", 
    "S1 + AMSR2 (All)"
]
bar_order = ["[6, 12, 18]", "[12, 24, 36]", "[18, 36, 54]"]
bar_color = ["lightsteelblue","cornflowerblue", "royalblue"]

fig, axes = plt.subplots(1, 4, figsize=(11,5))

for ax, col, ylabel, add_legend, yticklabels in zip(axes, 
                                                    ["SIC", "SOD", "Floe", "Combined"],
                                                    ["R2 (%)", "F1 (%)", "F1 (%)", "Combined (%)"],
                                                    [False, False, False, True],
                                                    [True, False, False, False],):
    df[col].unstack("Atrous Rate")[bar_order].plot.bar(ax=ax, 
                                                       legend=add_legend, 
                                                       ylim=[0,100], 
                                                       title=col,
                                                       color=bar_color,
                                                       xlabel='',
                                                       ylabel=ylabel,
                                                       )
    ax.grid(which="major", axis="y", ls="--", zorder=0)
    ax.set_xticklabels(xlabels)
    ax.set_yticks(np.arange(0,101,10))
    if not yticklabels:
        ax.set_yticklabels([])
    if add_legend:
        ax.legend(fontsize="small", loc="upper left", title="Atrous rates")
#ax.set_xlabel(xlabels)
```

```python
ylabels = [
    "AMSR2 (18.7, 36.5)", 
    "AMSR2 (18.7, 36.5, 23.8)", 
    "AMSR2 (All)", 
    "S1", 
    "S1 + AMSR2 (18.7, 36.5)", 
    "S1 + AMSR2 (All)"
]
bar_order = ["[6, 12, 18]", "[12, 24, 36]", "[18, 36, 54]"]
bar_color = ["lightsteelblue","cornflowerblue", "royalblue"]

fig, axes = plt.subplots(4, 1, figsize=(10,9))

for ax, col, xlabel, add_legend, xticklabels in zip(axes, 
                                                    ["SIC", "SOD", "Floe", "Combined"],
                                                    ["R2 (%)", "F1 (%)", "F1 (%)", "Combined (%)"],
                                                    [True, False, False, False],
                                                    [False, False, False, True],):
    df[col].unstack("Atrous Rate")[bar_order].plot.barh(ax=ax, 
                                                       legend=add_legend, 
                                                       xlim=[0,100],
                                                       color=bar_color,
                                                       ylabel='',
                                                       xlabel=xlabel,
                                                       )
    ax.grid(which="major", axis="x", ls="--", zorder=0)
    ax.set_yticklabels(ylabels)
    ax.set_xticks(np.arange(0,101,10))
    if not xticklabels:
        ax.set_xticklabels([])
    if add_legend:
        ax.legend(fontsize="small", loc="lower right", title="Atrous rates")
    ax.text(0.99, 0.98, col, horizontalalignment='right',
            verticalalignment='top', transform=ax.transAxes,
            fontsize=15)

fig.subplots_adjust(hspace=0.3)
#ax.set_xlabel(xlabels)
```
